```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import markers
import matplotlib.font_manager as font_manager
import matplotlib.colors
from pylab import rcParams

# Figures settings
%config InlineBackend.figure_format = 'jpg'
import matplotlib as mpl
import matplotlib.ticker as ticker
```

```python
bgData=pd.read_csv("/Users/nsha457/Documents/Jupyter_files/SBML_AnnotMerge/halfgloucose to AcCOA/BIOMD0000000017/Comparison/bgData.csv")
solutions=pd.read_csv("/Users/nsha457/Documents/Jupyter_files/SBML_AnnotMerge/halfgloucose to AcCOA/BIOMD0000000017/Comparison/solutions.csv")


# Computing NRMSE

RMSE_AcP=np.sqrt(sum([(a-b)**2 for a,b in zip(bgData[str(23)][0:500],solutions.AcP[0:500])])/len(bgData[str(23)][0:500]))
RMSE_percent_AcP=RMSE_AcP/(max(solutions.AcP[0:500])-min(solutions.AcP[0:500]))*100

RMSE_CoA=np.sqrt(sum([(a-b)**2 for a,b in zip(bgData[str(20)][0:500],solutions.CoA[0:500])])/len(bgData[str(20)][0:500]))
RMSE_percent_CoA=RMSE_CoA/(max(solutions.CoA[0:500])-min(solutions.CoA[0:500]))*100

RMSE_pyruvate=np.sqrt(sum([(a-b)**2 for a,b in zip(bgData[str(18)][0:500],solutions.pyruvate[0:500])])/len(bgData[str(18)][0:500]))
RMSE_percent_pyruvate=RMSE_pyruvate/(max(solutions.pyruvate[0:500])-min(solutions.pyruvate[0:500]))*100

RMSE_AcCoA=np.sqrt(sum([(a-b)**2 for a,b in zip(bgData[str(22)][0:500],solutions.AcCoA[0:500])])/len(bgData[str(22)][0:500]))
RMSE_percent_AcCoA=RMSE_AcCoA/(max(solutions.AcCoA[0:500])-min(solutions.AcCoA[0:500]))*100

RMSE_lactate=np.sqrt(sum([(a-b)**2 for a,b in zip(bgData[str(19)][0:500],solutions.lactate[0:500])])/len(bgData[str(19)][0:500]))

RMSE_halfglucose=np.sqrt(sum([(a-b)**2 for a,b in zip(bgData[str(21)][0:500],solutions.halfglucose[0:500])])/len(bgData[str(21)][0:500]))


# Plotting and comparing the simulations for the first model (pyruvate distribution)

mpl.rc('font',family='Times New Roman')
rcParams['figure.figsize'] = 10, 10

fig, axs = plt.subplots(3,2)
plt.subplots_adjust(hspace = 0.0 )
plt.subplots_adjust(top = 0.85, bottom=0.1, hspace=0.4, wspace=0.4)

subs=['Glucose','Lactate','AcCoA','Pyruvate','CoA','AcP']

axs[0,0].plot(solutions.Time[0:500], 1000*solutions.halfglucose[0:500], '#EF9F5B', linewidth=3, label='SBML')
axs[0,0].plot(solutions.Time[0:500], 1000*bgData[str(21)][0:500], 'k--', linewidth=1, label='BG')
legend = axs[0,0].legend(loc='upper right', shadow=True, fontsize='8')
axs[0,0].set_xlabel('Time (s)',fontname='Times New Roman', fontsize='10')
axs[0,0].set_ylabel('Concentration (uM)',fontname='Times New Roman', fontsize='10')  
axs[0,0].set_title(subs[0],fontsize=12, fontweight='bold')
axs[0,0].text(0.65,0.20, "(NRMSE= {:.1f} %)".format(RMSE_halfglucose), fontsize='10',fontweight='bold',transform=axs[0,0].transAxes)


axs[0,1].plot(solutions.Time[0:500], 1000*solutions.lactate[0:500], '#EF9F5B', linewidth=3)
axs[0,1].plot(solutions.Time[0:500], 1000*bgData[str(19)][0:500], 'k--', linewidth=1)
axs[0,1].set_xlabel('Time (s)',fontname='Times New Roman', fontsize='10')
axs[0,1].set_ylabel('Concentration (uM)',fontname='Times New Roman', fontsize='10')  
axs[0,1].set_title(subs[1],fontsize=12, fontweight='bold')
axs[0,1].text(0.65,0.20, "(NRMSE= {:.1f} %)".format(RMSE_lactate), fontsize='10',fontweight='bold',transform=axs[0,1].transAxes)


axs[1,0].plot(solutions.Time[0:500], 1000*solutions.AcCoA[0:500], '#EF9F5B', linewidth=3)
axs[1,0].plot(solutions.Time[0:500], 1000*bgData[str(22)][0:500], 'k--', linewidth=1)
axs[1,0].set_xlabel('Time (s)',fontname='Times New Roman', fontsize='10')
axs[1,0].set_ylabel('Concentration (uM)',fontname='Times New Roman', fontsize='10')  
axs[1,0].set_title(subs[2],fontsize=12, fontweight='bold')
axs[1,0].text(0.65,0.20, "(NRMSE= {:.1f} %)".format(RMSE_percent_AcCoA), fontsize='10',fontweight='bold',transform=axs[1,0].transAxes)


axs[1,1].plot(solutions.Time[0:500], 1000*solutions.pyruvate[0:500], '#EF9F5B', linewidth=3)
axs[1,1].plot(solutions.Time[0:500], 1000*bgData[str(18)][0:500], 'k--', linewidth=1)
axs[1,1].set_xlabel('Time (s)',fontname='Times New Roman', fontsize='10')
axs[1,1].set_ylabel('Concentration (uM)',fontname='Times New Roman', fontsize='10')  
axs[1,1].set_title(subs[3],fontsize=12, fontweight='bold')
axs[1,1].text(0.65,0.20, "(NRMSE= {:.1f} %)".format(RMSE_percent_pyruvate), fontsize='10',fontweight='bold',transform=axs[1,1].transAxes)


axs[2,0].plot(solutions.Time[0:500], 1000*solutions.CoA[0:500], '#EF9F5B', linewidth=3)
axs[2,0].plot(solutions.Time[0:500], 1000*bgData[str(20)][0:500], 'k--', linewidth=1)
axs[2,0].set_xlabel('Time (s)',fontname='Times New Roman', fontsize='10')
axs[2,0].set_ylabel('Concentration (uM)',fontname='Times New Roman', fontsize='10')  
axs[2,0].set_title(subs[4],fontsize=12, fontweight='bold')
axs[2,0].text(0.65,0.20, "(NRMSE= {:.1f} %)".format(RMSE_percent_CoA), fontsize='10',fontweight='bold',transform=axs[2,0].transAxes)


axs[2,1].plot(solutions.Time[0:500], 1000*solutions.AcP[0:500], '#EF9F5B', linewidth=3)
axs[2,1].plot(solutions.Time[0:500], 1000*bgData[str(23)][0:500], 'k--', linewidth=1)
axs[2,1].set_xlabel('Time (s)',fontname='Times New Roman', fontsize='10')
axs[2,1].set_ylabel('Concentration (uM)',fontname='Times New Roman', fontsize='10')  
axs[2,1].set_title(subs[5],fontsize=12, fontweight='bold')
axs[2,1].text(0.65,0.20, "(NRMSE= {:.1f} %)".format(RMSE_percent_AcP), fontsize='10',fontweight='bold',transform=axs[2,1].transAxes)


plt.savefig('firstModelApproximation.jpg', dpi=300)
```
